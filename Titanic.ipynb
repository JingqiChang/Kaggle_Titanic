```python
from sklearn.ensemble import RandomForestRegressor

def set_missing_age(df):
    age_df=df[['Age','Fare','Parch','SibSp','Pclass']]
    known_age=age_df[age_df.Age.notnull()].values
    unknown_age=age_df[age_df.Age.isnull()].values
    
    y=known_age[:,0]
    x=known_age[:,1:]
    
    rfr=RandomForestRegressor(random_state=0,n_estimators=2000,n_jobs=-1)
    rfr.fit(x,y)
    
    predictedAges=rfr.predict(unknown_age[:,1:])
    df.loc[(df.Age.isnull()),'Age']=predictedAges
    return df

train=set_missing_age(train)
```

```python
data=train.drop(train[train.Embarked.isnull()].index)
```

```python
df=data[['Pclass','Sex','Age','SibSp','Parch','Fare','Survived']]
cor=df.corr()
print(cor)

plt.figure(figsize=(12,10))
sns.heatmap(cor,cmap='RdYlGn',center=0,annot=True)
plt.show()
```

```python
plt.figure(figsize=(10,8),dpi=100)

p1=plt.subplot(221)
Pclass_0=data.loc[data['Survived']==0,'Pclass'].value_counts()
Pclass_1=data.loc[data['Survived']==1,'Pclass'].value_counts()
df_pclass=pd.DataFrame({'survived':Pclass_1,'unsurvived':Pclass_0})
df_pclass.plot(kind='bar',ax=p1)
plt.xticks(rotation=0)
plt.title('Survived in different class')

p2=plt.subplot(222)
Survive_m=data.loc[data['Sex']=='male','Survived'].value_counts()
Survive_f=data.loc[data['Sex']=='female','Survived'].value_counts()
df_sex=pd.DataFrame({'male':Survive_m,'female':Survive_f})
df_sex.plot(kind='bar',stacked=True,ax=p2)
plt.xticks(rotation=0)
plt.title('Survived in different sex')

p3=plt.subplot(223)
bin=[min(data['Age']),10,60,max(data['Age'])]
bin_label=['under 10','10-60','over 60']
data['Age_cut']=pd.cut(data['Age'],bins=bin, labels=bin_label)
Age_0=data.loc[data['Survived']==0,'Age_cut'].value_counts()
Age_1=data.loc[data['Survived']==1,'Age_cut'].value_counts()
df_age=pd.DataFrame({'Survived':Age_0,'Unsurvived':Age_1})
df_age.plot(kind='bar',ax=p3)
plt.xticks(rotation=45)
plt.title('Survived in different Age')

p4=plt.subplot(224)
Embarked_0=data.Embarked[data.Survived==0].value_counts()
Embarked_1=data.Embarked[data.Survived==1].value_counts()
df_Embarked=pd.DataFrame({'Survived':Embarked_0,'Unsurvived':Embarked_1})
df_Embarked.plot(kind='bar',stacked=True,ax=p4)
plt.xticks(rotation=0)
plt.title('Survived in different embark')

plt.show()
```
